# Probability Basics

For *very large number of trials or observations*, Probability tells us the expected percentage of times a particular outcome will occur (materialize). Consider the coin flip experiment - the Probability tends to 50% as number of coin flips grows towards large numbers. In the illustration below, we consider a 'Head' as 0.5, and 'Tail' as 0.5. We then generate random drawws of verying sizes, and sum them to determine how Heads and Tails are distributed. The bar graph plots the frequencies of different experiments.

Note that as 'num_trials' is increased, the bar graph becomes more concentrated at '0' - indicating that a large number of experiments result in equi-likely Heads and Tails

```python
import random
import numpy as np
from collections import Counter 
import matplotlib.pyplot as plt

num_trials = 1000

print("Generating", num_trials, "random coin flip trials")

exp_list = np.array([], dtype=float)

for n in range(0, num_trials):
    # Generate a list of 'num_trials' random integers in [0, 2) using numpy and calculate its sum, and append to exp_list
    exp_list = np.append(exp_list, sum(np.random.randint(0, 2, num_trials) - 0.5))

exp_hist = Counter(exp_list)

fig, ax = plt.subplots()
g1 = ax.bar(exp_hist.keys(), exp_hist.values())
plt.show()
```
